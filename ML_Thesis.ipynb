```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile, f_classif
```

```python
# The solution to categorical variables is to use dummy coding before passing to the model
categorical_columns=['protocol_type', 'service', 'flag']
data_categorical_values = data_set[categorical_columns]
data_categorical_values.head()
```

```python
# Make column names for dummies
# protocol type
unique_protocol = sorted(data_set.protocol_type.unique())
string1 = 'Protocol_type_'
unique_protocol2=[string1 + x for x in unique_protocol]
# service
unique_service=sorted(data_set.service.unique())
string2 = 'service_'
unique_service2=[string2 + x for x in unique_service]
# flag
unique_flag=sorted(data_set.flag.unique())
string3 = 'flag_'
unique_flag2=[string3 + x for x in unique_flag]
# put together
dumcols=unique_protocol2 + unique_service2 + unique_flag2
print(dumcols)
```

```python
# Transform categorical variables to numerics
data_categorical_values_enc = data_categorical_values.apply(LabelEncoder().fit_transform)
print(data_categorical_values_enc.head())
data_categorical_values.describe()
```

```python
enc = OneHotEncoder()
data_binary_encoder = enc.fit_transform(data_categorical_values_enc)
data_cat_data = pd.DataFrame(data_binary_encoder.toarray(),columns=dumcols)
data_cat_data.head()
```

```python
# Join encoded data frame with numeric data frame
newdf = data_set.join(data_cat_data)
newdf.drop('flag', axis=1, inplace=True)
newdf.drop('protocol_type', axis=1, inplace=True)
newdf.drop('service', axis=1, inplace=True)
print(newdf.shape)
```

```python
# Take label column
labeldf = newdf['label']
# Change the label column(0 -> good connection, 1 -> bad connection)
newlabeldf = labeldf.replace({ 'normal.' : 0, 'neptune.' : 1 ,'back.': 1, 'land.': 1, 'pod.': 1, 'smurf.': 1, 'teardrop.': 1,'mailbomb.': 1, 'apache2.': 1, 'processtable.': 1, 'udpstorm.': 1, 'worm.': 1,
                              'ipsweep.' : 1,'nmap.' : 1,'portsweep.' : 1,'satan.' : 1,'mscan.' : 1,'saint.' : 1
                              ,'ftp_write.': 1,'guess_passwd.': 1,'imap.': 1,'multihop.': 1,'phf.': 1,'spy.': 1,'warezclient.': 1,'warezmaster.': 1,'sendmail.': 1,'named.': 1,'snmpgetattack.': 1,'snmpguess.': 1,'xlock.': 1,'xsnoop.': 1,'httptunnel.': 1,
                              'buffer_overflow.': 1,'loadmodule.': 1,'perl.': 1,'rootkit.': 1,'ps.': 1,'sqlattack.': 1,'xterm.': 1})
# Put the new label column back
newdf['label'] = newlabeldf
print(newdf['label'].head())
```

```python
# Split dataframes into X & Y
# Y_train as the label column
Y2_train = newdf['label']
```

```python
# Check Y_train
Y2_train.shape
Y2_train.head
```

```python
# Grab info
print(Y2_train.value_counts())
```

```python
# Create X_train as a dataframe of features
X_train = newdf.drop('label', 1)
X_train.shape
```

```python
colNames=list(X_train)
```

```python
# Scale the dataframe
scaler1 = preprocessing.StandardScaler().fit(X_train)
X_train = scaler1.transform(X_train) 
```

```python
# Feature Selection
# univariate feature selection with ANOVA F-test. using secondPercentile method
# Scikit-learn exposes feature selection routines as objects that implement the transform method
# SelectPercentile: removes all but a user-specified highest scoring percentage of features
# f_classif: ANOVA F-value between label/feature for classification tasks

from sklearn.feature_selection import SelectPercentile, f_classif
np.seterr(divide='ignore', invalid='ignore');
selector=SelectPercentile(f_classif, percentile=10)
X_new = selector.fit_transform(X_train,Y2_train)
X_new.shape
```

```python
# Get the features
true = selector.get_support()
newcolindex = [i for i, x in enumerate(true) if x]
newcolname = list( colNames[i] for i in newcolindex )
newcolname
```

```python
print('Features selected:', newcolname)
print()
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
# Create a decision tree classifier. By convention, clf means 'classifier'
clf = DecisionTreeClassifier(random_state=0)

#rank all features, i.e continue the elimination until the last one
rfe = RFE(clf, n_features_to_select=1)
rfe.fit(X_new, Y2_train)
print ("DoS Features sorted by their rank:")
print (sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), newcolname)))
```

```python
final_data = newdf[['logged_in',
                     'count',
                     'srv_count',
                     'srv_diff_host_rate',
                     'dst_host_count',
                     'dst_host_same_src_port_rate',
                     'Protocol_type_icmp',
                     'Protocol_type_tcp',
                     'service_ecr_i',
                     'service_http',
                     'service_smtp',
                     'label' ]]
```

```python
# Split dataset in features and target variable
feature_cols = ['logged_in',
                'count',
                'srv_count',
                'srv_diff_host_rate',
                'dst_host_count',
                'dst_host_same_src_port_rate',
                'Protocol_type_icmp',
                'Protocol_type_tcp',
                'service_ecr_i',
                'service_http',
                'service_smtp',
                ]
X = final_data[feature_cols] # Features
y = final_data.label # Target variable
```

```python
# Split dataset into training set and test set
training_data, testing_data, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) # 80% training and 20% test
```

```python
training_data
```

```python
testing_data
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(training_data,y_train)

# Predict the response for test dataset
Decision_Trees_pred = clf.predict(testing_data)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn import metrics

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

print("F1 Score: {}".format(f1_score(y_test, Decision_Trees_pred)))
print("Recall score: {}".format(recall_score(y_test, Decision_Trees_pred)))
print("Precision score: {}".format(precision_score(y_test, Decision_Trees_pred)))
print("Accuracy:",metrics.accuracy_score(y_test, Decision_Trees_pred))
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, Decision_Trees_pred)
cnf_matrix
```

```python
import joblib
import os
joblib.dump(clf, "Decision_trees.joblib") 
print(f"Decision Trees size: {np.round(os.path.getsize('Decision_trees.joblib')/1024 / 1024,2) } MB")
```
